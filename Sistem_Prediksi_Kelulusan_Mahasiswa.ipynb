```python
import pandas as pd

# Memuat data
file_path = '/content/Data_Training_Dummy.csv'
data = pd.read_csv(file_path)

# Melihat beberapa baris pertama dari data
print(data.head())

# Informasi umum tentang data
print(data.info())

# Statistik deskriptif
print(data.describe())

# Melihat distribusi kelas target
print(data['Hasil'].value_counts())
```

```python
data_rounded = data.round(2)

# Menyimpan data yang sudah dibulatkan ke file baru
data_rounded.to_csv('Data_Training_Dummy_Rounded.csv', index=False)

# Menampilkan data yang sudah dibulatkan
print(data_rounded.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Mengatasi nilai yang hilang (jika ada)
data_rounded = data_rounded.dropna()

# Encoding fitur kategorikal
label_encoders = {}
categorical_columns = data_rounded.select_dtypes(include=['object']).columns

for column in categorical_columns:
    le = LabelEncoder()
    data_rounded[column] = le.fit_transform(data_rounded[column])
    label_encoders[column] = le

# Memisahkan fitur dan target
X = data_rounded.drop('Hasil', axis=1)
y = data_rounded['Hasil']
```

```python
data_rounded.head(10)
```

```python
from sklearn.model_selection import train_test_split

# Membagi data menjadi set pelatihan dan set pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Memeriksa ukuran dari masing-masing set dan distribusi kelas
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
print(y_train.value_counts(), y_test.value_counts())
```

```python
from sklearn.naive_bayes import GaussianNB

# Inisialisasi model
model = GaussianNB()

# Melatih model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Memprediksi nilai menggunakan set pengujian
y_pred = model.predict(X_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi: {accuracy}')

# Laporan klasifikasi
print(classification_report(y_test, y_pred))
```

```python
import joblib

# Menyimpan model ke file
model_path = 'model3.pkl'
joblib.dump(model, model_path)
```
